```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.interpolate
from scipy import stats
import numpy as np

var = pd.read_excel("Book1.xlsx")
print(var)

fig = plt.figure()
x = list(var['X values'])
y = list(var['Y values'])
#define data
X = np.array(x)
Y = np.array(y)

#find line of best fit
a, b = np.polyfit(X, Y, 1)

#add points to plot


#add line of best fit to plot
plt.title("Plot of variation of Johnson Noise Vs. Resistance",fontweight='bold')
plt.plot(X, a*X+b,'o-',color = 'black',label='Best Fitted Line')  
plt.text(1,19,"Equation for Best Fitted Line",**csfont, size=12)
plt.text(1, 17, '$y$ = ' + '{:.6f}'.format(a)+ '$x$' + ' + {:.2f}'.format(b) , size=12)
#create plot of x vs. y
plt.plot(x, y, '--Xr', label='Interpolation Line')
plt.xlabel("Resistance ($\Omega$)",fontweight='bold',fontsize=12)
plt.ylabel(  "Johnson Noise $(V_J ^{2} ~x~10^{-12}$ in ${Volt}^{2})$",fontweight='bold',fontsize=12)
plt.legend()
plt.grid()
plt.show()
fig.savefig('plt1.pdf')
```
